Q:what is percentage of crop yield of different states?

```python
plt.figure(figsize=(12,6))
sns.countplot(data=ag,x="State")
plt.title("crop Yield of different states")
plt.xticks(rotation=90)
```

```python
Q:which variety is most cultivating?
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=ag,x=ag["Variety"].head(20),y='Max Price')
plt.xticks(rotation=80)
```

```python
ag['Variety'].value_counts()
```

```python
what is count of the Grade?
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=ag,x='Grade')
```

Q:WHat is the maximum price in each state?

```python
plt.figure(figsize=(12,6))
sns.barplot(data=ag,x="State",y="Max Price")
plt.xticks(rotation=60)
plt.title("Max price in each state")
```

```python
Q:what is the modal price in each state?
```

plt.figure(figsize=(12,6))

```python
plt.figure(figsize=(12,6))
sns.barplot(data=ag,x=ag['Variety'].head(20),y='Max Price')
plt.xticks(rotation=90)
```

Q:what are the top 30 commodities?

```python
sns.histplot(data=ag,x=ag['Commodity'].head(30),kde=True)
plt.xticks(rotation=80)
```

Variations of Prize distribution?

```python
x=ag[['Min Price',"Max Price","Modal Price"]]
plt.figure(figsize=(6,6))
sns.boxplot(data=x)
plt.title("Prize distribution")
```

Q:show the max price in each state in scatter plot?

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=ag,x='State',y="Max Price")
plt.xticks(rotation=80)
```

```python
top_data=ag.nlargest(10,"Max Price")
```

What are the commodities in top countries?

```python
plt.figure(figsize=(12,6))
sns.barplot(data=top_data,x="Max Price",y="State",hue="Commodity")
```

```python
small_data=ag.nsmallest(10,'Modal Price')
```

Q:what is the commodities in the least countries ?

```python
plt.figure(figsize=(12,8))
sns.barplot(data=small_data,x='Max Price',y='State',hue='Commodity')
```
