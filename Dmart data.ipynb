```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dm=pd.read_csv("Dmart.csv")
```

```python
dm.isna().sum()
```

```python
dm['Name']=dm['Name'].fillna("Kitchen Appliance")
dm['Brand']=dm['Brand'].fillna('Local/Unknown')
dm[dm['Price'].isnull()==True]
```

```python
dm['Category'] = dm['Category'].fillna('Home & Kitchen')
dm['SubCategory'] = dm['SubCategory'].fillna('Home Appliances')
dm['BreadCrumbs']=dm['BreadCrumbs'].fillna('Home & Kitchen > Cookware & Serveware')
dm[dm.Quantity.isnull()==True]
```

```python
dm['Quantity'] = dm['Quantity'].fillna('Size XXL')
dm['Description'] = dm['Description'].fillna('No Desciption')
```

```python
dm['Price']=dm['Price'].fillna(dm['Price'].mean())
```

```python
dm['DiscountedPrice']=dm['DiscountedPrice'].fillna(dm['DiscountedPrice'].mean())
```

```python
dm.isna().sum()
```

```python
dm['Category'].value_counts()
```

what are the number of categories present in dmart?

```python
plt.figure(figsize=(12,5))
sns.countplot(data=dm,x='Category')
plt.xticks(rotation=90)
plt.title("Number of categorires")
```

what are the diffrent prices of  the categories in dmart?

```python
plt.figure(figsize=(12,5))
sns.barplot(data=dm,x='Category',y='Price')
plt.xticks(rotation=90)
plt.title("prices of the category")
```

```python
most=dm["Category"].sort_values(ascending=False)
```

what are the top categories present in dmart?

```python
most.value_counts().plot(kind='bar',color=sns.color_palette(),figsize=(10,5))
plt.title("top categories in the dmart")
plt.show()
```
