#        EXPLORATORY DATA ANALYSIS ON MOBILES

# Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('smartphone.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df['rating']=df['rating'].fillna(0)
```

```python
df['processor_brand']=df['processor_brand'].fillna('unknown')
```

```python
df['fast_charging'].unique()
```

```python
df['battery_capacity']=df['battery_capacity'].fillna(1831)
```

```python
df['fast_charging']=df['fast_charging'].fillna(22)
```

```python
df['extended_upto'].unique()
```

```python
df['extended_upto']=df['extended_upto'].fillna(32.)
```

```python
df.isna().sum()
```

# Data Visualiozation

```python
df.head()
```

```python
df['brand_name'].value_counts()
```

what are the types of brands present in mobiles?

```python
plt.figure(figsize=(12,5))
sns.countplot(data=df,x='brand_name')
plt.xticks(rotation=90)
plt.title("brand of the mobiles")
```

what is the price of mobiles as per brand names?

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df,x='brand_name',y='price')
plt.xticks(rotation=60)
plt.title('prices of the brands')
```

Q:top 10 brands in the mobile

```python
df['brand_name'].value_counts().head(10).plot(kind='bar',figsize=(12,6),color=sns.color_palette())
plt.title('top 10 brands in mobile')
```

Q: top 10 brands in the pie chart?

```python
df['brand_name'].value_counts().head(10).plot(kind='pie',figsize=(12,6),autopct="%.2f%%")
plt.title('top 10 brands in mobile')
```

```python
df['battery_capacity'].unique()
```

what are the prices of the batteries as per capacity?

```python
plt.figure(figsize=(14,4))
sns.barplot(data=df,x='battery_capacity',y="price")
plt.xticks(rotation=90)
plt.title('price of battery capacities ')
```

show the distribution of ram in the pie chart?

```python
df['extended_upto'].value_counts().plot(kind='pie',autopct="%.2f%%",figsize=(12,6))
plt.title("distribution of ram capacity in mobiles")
```

what is the prices of the rams?

```python
plt.figure(figsize=(14,4))
sns.barplot(data=df,x='extended_upto',y="price")
plt.xticks(rotation=90)
plt.title('price of rams')
```
