# importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# importing the dataset

```python
lf=pd.read_csv("layoffs_data.csv")
```

```python
lf.head()
```

#basic info of the dataset

```python
lf.info()
```

# checking the null values in the dataset

```python
lf.isna().sum()
```

```python
lf['Percentage'].value_counts()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(data=lf,x="Industry",y='Funds_Raised')
plt.xticks(rotation=90)
```

```python
lf['Location_HQ'].shape
```

```python
Location=lf.groupby(['Location_HQ']).size()
Location=Location.sort_values(ascending=False)
```

# Q:in which country there are most head quarters

```python
Location.head(20).plot(kind='bar',figsize=(12,6),color=sns.color_palette())
```

# what are the number stages in the dataset?

```python
sns.countplot(data=lf,x='Stage')
plt.xticks(rotation=90)
plt.title("The number of Stages in the Dataset")
```

```python
lf['Company'].value_counts()
```

# what are the company stages in the dataset?

```python
sns.countplot(data=lf,x=lf['Company'].head(40))
plt.xticks(rotation=90)
plt.title("The number of Stages in the Dataset")
```

```python
lf.head()
```

# grouped the Funds Raised in the dataset?

```python
grouped=lf.nlargest(20,'Funds_Raised')
```

```python
grouped
```

Q:which are the largest fund raised company?

```python
plt.figure(figsize=(12,6))
sns.barplot(data=grouped,x="Company",y="Funds_Raised")
```

what are the places that the Head Quaters are located in the dataset?

```python
plt.figure(figsize=(12,5))
sns.countplot(data=lf,x=lf["Location_HQ"].head(20))
```

```python
lf['Location_HQ'].value_counts()
```

```python
Location=lf.groupby(['Location_HQ']).size()
```

```python
a=Location.sort_values(ascending=False)
a.head()
```

what are the top 10 FUnds Raised companies?

```python
a.head(10).plot(kind='bar',x="Funds_Raised",figsize=(12,6),color=sns.color_palette())
plt.show()
```

what are the least 10 head quaters located cities?

```python
a.tail(10).plot(kind='bar',figsize=(12,5),color=sns.color_palette())
```
