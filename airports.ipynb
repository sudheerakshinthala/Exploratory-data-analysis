# World Airports Data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
air=pd.read_csv("airports.csv")
```

```python
air
```

```python
air.head()
```

```python
air.info()
```

```python
air.isna().sum()
```

```python
air['continent']=air['continent'].fillna('unknown')
```

```python
air['iso_country']=air['iso_country'].fillna('unknown')
```

```python
air.drop(['home_link', 'wikipedia_link', 'keywords', 'iata_code'], axis=1, inplace=True)
```

```python
air.isna().sum()
```

# Summary statistics of numerical columns

```python
air.describe()
```

```python
air.columns
```

# Distribution of Elevation

```python
plt.figure(figsize=(8, 6))
sns.histplot(air['elevation_ft'], kde=True)
plt.title('Distribution of Elevation')
plt.xlabel('Elevation (ft)')
plt.ylabel('Count')
plt.show()
```

# Number of Airports by Type

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=air, x='type')
plt.title('Number of Airports by Type')
plt.xlabel('Type')
plt.xticks(rotation=90)
plt.ylabel('Count')
plt.show()
```

# Percentage of Airports with Scheduled Service

```python
plt.figure(figsize=(8, 6))
air['scheduled_service'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Percentage of Airports with Scheduled Service')
plt.show()
```

# Elevation Distribution by Airport Type

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=air, x='type', y='elevation_ft')
plt.title('Elevation Distribution by Airport Type')
plt.xlabel('Type')
plt.ylabel('Elevation (ft)')
plt.xticks(rotation=45)
plt.show()
```

# Top 10 Countries with the Highest Number of Airports

```python
top_countries = air['iso_country'].value_counts().nlargest(10)
plt.figure(figsize=(10, 8))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title('Top 10 Countries with the Highest Number of Airports')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Geographical Distribution of Airports/Heliports

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data=air, x='longitude_deg', y='latitude_deg', hue='type', alpha=0.7)
plt.title('Geographical Distribution of Airports/Heliports')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

# Relationship between Scheduled Service and Airport Type

```python
plt.figure(figsize=(10, 8))
sns.countplot(data=air, x='type', hue='scheduled_service')
plt.title('Relationship between Scheduled Service and Airport Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Scheduled Service')
plt.show()
```

# Elevation Distribution by Airport/Heliport Type

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=air, x='type', y='elevation_ft')
plt.title('Elevation Distribution by Airport/Heliport Type')
plt.xlabel('Type')
plt.ylabel('Elevation (ft)')
plt.xticks(rotation=45)
plt.show()
```

# Relationship between Scheduled Service and Airport Type

```python
plt.figure(figsize=(10, 8))
sns.countplot(data=air, x='type', hue='scheduled_service')
plt.title('Relationship between Scheduled Service and Airport Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Scheduled Service')
plt.show()
```

# What are the airport types in dataset

```python
air_type = air.type.value_counts()
plt.figure(figsize=(12, 6))
plt.title("Airport types", fontsize=16)
plt.xlabel("Quantity", fontsize=12)
plt.ylabel("Type", fontsize=12)
sns.barplot(x=air_type.values, y=air_type.index, palette='plasma')
plt.show()
```

# Conclusion and Insights
In this exploratory data analysis (EDA) of the airports and heliports dataset, we have gained valuable insights into the distribution, characteristics, and relationships within the data. Here are some key findings:

1. Distribution of Airport/Heliport Types: The dataset contains various types of facilities, with small airports being the most prevalent, followed by heliports. Closed, seaplane base, balloonport, medium airports, and large airports are less common.

2. Elevation Distribution: The elevation of airports and heliports varies widely. While most facilities have elevation information available, some entries lack this data. The box plot analysis revealed variations in elevation based on the type of facility, with large airports tending to have higher elevations compared to other types.

3. Geographical Distribution: The scatter plot and geographical analysis showcased the global distribution of airports and heliports. Concentration of facilities is observed in North America (particularly the United States), Brazil, Japan, Canada, and Australia.

4. Scheduled Service and Facility Types: The relationship between scheduled service and facility types indicated that most airports and heliports do not have scheduled commercial service. However, certain types, such as medium airports and large airports, are more likely to offer scheduled service.

5. Top Countries with the Highest Number of Airports/Heliports: The analysis identified the top countries with the highest number of airports and heliports. The United States leads with the most facilities, followed by Brazil, Japan, Canada, and Australia.
