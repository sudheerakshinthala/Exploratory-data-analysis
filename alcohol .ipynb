# First 5 rows in the dataset

```python
mh.head()
```

```python
grouped1= mh.groupby(['Countries']).size()
grouped1.head(10)
```

```python
most=grouped1.sort_values(ascending=False)
most.head(10)
```

# Data visualization

Most alocohol consumption countries?

```python
most.head(15).plot(kind='pie',autopct="%.2f%%",figsize=(12,6))
plt.title("alcohol consumption of countries in percentage")
```

```python
grouped = mh.groupby(['Date']).size()
grouped.head()
```

what is the top 10 dates of alcohol consumption?

```python
grouped.head(10).plot(kind='bar',color=sns.color_palette())
plt.xticks(rotation=90)
plt.title("top 10 dates of alcohol consumption")
```

what is the least 10 dates of alcohol consumption?

```python
grouped.tail(10).plot(kind='bar',color=sns.color_palette())
plt.xticks(rotation=90)
plt.title("least 10 dates of alcohol consumption")
```

what is the top 10 gender wise alcohol consumption in dataset?

```python
plt.figure(figsize=(12,5))
sns.countplot(data=mh,x=mh['Countries'].head(20),hue="Gender")
plt.xticks(rotation=90)
```

what is the alcohol consumption of male vs female?

```python
mh["Gender"].value_counts().plot(kind='pie',autopct="%.2f%%")
plt.title("alcohol consumption male vs female")
```

```python
mh.tail()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=mh,x="Count",y=mh["Countries"].head(20))
```

```python
most.head(40).plot(kind='line',marker='o',figsize=(10,4))
plt.title("most alocohol consumpties countries")
```

# changed the datatype of date (object)to the datetime datatype

```python
mh['Date']=pd.to_datetime(mh["Date"])
```

```python
top_data=mh.nlargest(12,"Date")
```

alcohol consumption by the countries in Top 10 dates?

```python
sns.countplot(data=top_data,x="Countries")
plt.xticks(rotation=90)
plt.title("alcohol consumption by the countries in Top 10 dates")
```
