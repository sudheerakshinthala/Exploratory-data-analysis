# DATA VISUALIZATION

```python
re.head()
```

# what are the types present in Status

```python
plt.figure(figsize=(12,6))
sns.countplot(data=re,x='Status')
plt.title('The types present in status')
```

what is the status in jnv?

```python
plt.figure(figsize=(12,3))
sns.barplot(data=re,x="JNV",y='Status')
plt.title("the status in the JNV")
plt.show()
```

what is the status in the kv?

```python
plt.figure(figsize=(12,3))
sns.barplot(data=re,x="KV",y='Status')
plt.title("the status in the KV")
```

```python
re.tail()
```

show the comparison of status in pie chart?

```python
re['Status'].value_counts().plot(kind='pie',autopct="%.2f%%")
plt.title("n=comparison in status")
```

what is the count of CTSA?

```python
plt.figure(figsize=(12,6))
sns.barplot(data=re,y='CTSA')
plt.xlabel("CTSA")
plt.title("The data of CTSA")
```

what is the analysis of GOVT AIDED?

```python
plt.figure(figsize=(12,6))
sns.barplot(data=re,y='GOVT AIDED')
plt.xlabel("govt aided")
plt.title('Analysis of Govt aided')
```

```python
re['INDEPENDENT'].value_counts()
```

what is the analysis of independemt?

```python
plt.figure(figsize=(12,4))
sns.barplot(data=re,y='INDEPENDENT')
plt.xlabel('Independent')
plt.title("Analysis of INdependent")
```

show the comparison between jNV and KV in pie chart?

```python
re[['JNV',"KV"]].value_counts().plot(kind='pie',autopct="%.2f55",figsize=(12,6))
plt.title("comparison of JNV and KV")
plt.show()
```

compare between INDEPENDENT and GOVT?

```python
re[['INDEPENDENT','GOVT']].plot(kind='bar',figsize=(12,6))
plt.title("INDEPENDENT VS GOVT")
plt.show()
```

compare the DATA between independent and govt aided?

```python
re[['INDEPENDENT','GOVT AIDED']].plot(kind='bar',figsize=(12,6))
plt.title("INDEPENDENT VS GOVT AIDED")
plt.show()
```

What is the status in the government?

```python
plt.figure(figsize=(12,4))
sns.barplot(data=re,y='GOVT',x='Status')
plt.title('Status in the government')
plt.show()
```

comparison of INDEPENDENT VS GOVT VS GOVT AIDED in boxplot?

```python
a=re[['INDEPENDENT','GOVT','GOVT AIDED']]
plt.figure(figsize=(12,5))
sns.boxplot(data=a)
plt.title("comparison of INDEPENDENT VS GOVT VS GOVT AIDED")
plt.show()
```

show the Comparison between JNV and KV ?

```python
re[['JNV','KV']].plot(kind='line',marker='o')
plt.title("Comparison between JNV and KV")
plt.show()
```
